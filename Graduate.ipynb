```python
import pandas as pd

# Load the dataset from a CSV file
dataset = pd.read_csv("dataset.csv")

# Display the dataset
dataset
```

```python
# Extracting values from the dataset excluding the last column
x = dataset.iloc[:, :-1].values

# Display the extracted values
x
```

```python
# Extracting values from the last column of the dataset
y = dataset.iloc[:, -1].values

# Display the extracted values
y
```

```python
from sklearn.model_selection import train_test_split

# Splitting the dataset into training and testing sets
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2)
```

```python
# Checking the shape of the testing set
x_test.shape
```

```python
# Checking the shape of the training set
x_train.shape
```

```python
# Checking the shape of the testing labels
y_test.shape
```

```python
# Checking the shape of the training labels
y_train.shape
```

```python
print(y_train)
```

```python
#Regression model - Linear Regression
from sklearn.linear_model import LinearRegression
#create an object for linear regression class
regressor=LinearRegression()
regressor.fit(x_train,y_train)
```

```python
from sklearn.metrics import r2_score
y_pred=regressor.predict(x_test)
print(r2_score(y_pred,y_test)*100)
```
